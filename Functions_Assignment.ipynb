1.  What is the difference between a function and a method in Python?

- In Python, both functions and methods are callable entities that perform actions or computations.
- However, they differ primarily in how they are defined and used. Here's a detailed comparison:
- Function:
  - A function is defined using the def keyword outside of any class.
  - It is not tied to any specific object or data structure.
- Method:
  - A method is also defined using the def keyword but is within the body of a class.
  - It is tied to a specific object (instance) of the class or the class itself.
- Binding
  - Function:
   - A function is standalone and not associated with any object.
   - It can be called independently.
  - Method:
    - A method is bound to an instance or a class.
    - It requires an instance (or class for class methods) to be called, except for static methods.
-  Parameters
  - Function:
    - Functions take only the parameters explicitly defined in their signature.
  - Method:
    - Instance methods implicitly take the instance (self) as the first parameter.
- Type
  - Function:
    - Functions are objects of type function.
  - Method:
    - Methods are objects of type method.
- Scope
  - Function:
    - Functions can exist anywhere in a module and are typically used for general-purpose tasks.
  - Method:
    - Methods are scoped within a class and typically operate on data encapsulated by that class.

6.  Explain the concept of generators in Python and how they are defined.

- Generators are a special type of iterable in Python, used to generate values lazily (one at a time) as they are needed, rather than computing and storing all values in memory at once. They are particularly useful for handling large datasets or streams of data efficiently.

- Key Features of Generators:
  - Lazy Evaluation: Generators produce items on the fly and do not store the entire sequence in memory.
  - State Retention: Generators retain their state between iterations, making them suitable for tasks like streaming data.
  - Iterables: Generators are iterables themselves and can be used in a for loop or with functions like next().
- Generators can be defined using:
  -  Using yield in a Function
    - A function that uses the yield keyword becomes a generator function.
    - Each time yield is called, the generator produces a value and pauses its execution, resuming from the same point when the next value is requested.
 - Using Generator Expressions
   -  Similar to list comprehensions, but enclosed in parentheses () instead of brackets [].
    - Generator expressions are concise and memory-efficient.
- Advantages of Generators:
  - Memory Efficiency: Generate values on demand, avoiding the need to store large data structures in memory.
  - Simpler Code: Easy to implement complex iteration logic compared to manually managing iterators.
  - Performance: Faster for scenarios where only a subset of values is required.
- next(generator): Fetches the next value from the generator.
- StopIteration: Raised when the generator has no more values to produce.
- for Loop: Automatically handles StopIteration and iterates through all values.
- Generators are a powerful tool for creating iterators in Python, ideal for working with large or infinite sequences.
- They are defined using yield in a function or with generator expressions.
- Generators provide an efficient and elegant way to handle data lazily.

  Generators produce values lazily, one at a time, without storing the entire sequence in memory, making them ideal for handling large datasets or infinite sequences.

8.  What is a lambda function in Python and when is it typically used?

- A lambda function is a small, anonymous function defined using the lambda keyword. It can have any number of arguments but only a single expression. The expression's result is implicitly returned.
- When to Use Lambda Functions:
  - Short and Simple Functions:
    - Ideal for creating concise, one-time-use functions without defining a full function using def.
  - Functional Programming:
    - Commonly used with functions like map(), filter(), and reduce().
  - Sorting with Custom Keys: Used to define sorting logic inline.
  - Quick Inline Operations:
    - In situations where a function is used temporarily in the same scope.
- Limitations:
  - Restricted to a single expression (no statements or multiple lines).
  - Limited readability for complex logic compared to regular functions.
- Lambda functions are useful for creating quick, throwaway functions for simple tasks, particularly in functional programming or when brevity is preferred.

9. Explain the purpose and usage of the `map()` function in Python
- The map() function applies a given function to every item in an iterable (e.g., list, tuple) and returns a map object (an iterator) containing the results.
- Its purpose is to perform transformations or computations on all elements of an iterable without writing an explicit loop.
- Lazy Evaluation: map() returns an iterator; to get results, you need to convert it to a list, tuple, etc.
- Readable and Concise: Eliminates the need for verbose loops in many cases.
- Works with Multiple Iterables: Iterates over all provided iterables in parallel.
- Less readable for complex logic compared to list comprehensions.
- Requires converting the map object to a list or other structure for visibility.
- The map() function simplifies element-wise operations on iterables by applying a function to each item, making code concise and functional.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- These three functions are built-in functions in Python that are commonly used for functional programming tasks. They all operate on iterables but differ in how they process the data.
- map():
  - Purpose: Applies a given function to each item of an iterable (or iterables) and returns a map object (iterator) with the results.
  - Output: An iterable (map object) where each element is the result of applying the function to the corresponding item in the input iterable(s).
  - Usage: Used when you want to transform or modify each element of an iterable.
- reduce():
  - Purpose: Applies a function cumulatively to the items of an iterable, from left to right, reducing it to a single result.
  - Output: A single value resulting from applying the function repeatedly to the iterable.
  - Usage: Used for accumulating or combining elements, e.g., summing numbers, finding the product, or concatenating strings.
- filter():
  - Purpose: Filters elements from an iterable based on a function that returns True or False. Only elements that return True are included in the result.
  - Output: An iterable (filter object) containing only the elements that satisfy the condition specified by the function.
  - Usage: Used when you want to filter out elements based on a condition.
- map(): Used for transforming each item in an iterable.
- reduce(): Used for reducing an iterable to a single value through cumulative application of a function.
- filter(): Used for filtering out elements that don't meet a specific condition.

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is prime")
```

```python
# 5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration

        fibonacci_number = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fibonacci_number


fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

```python
# 7  Implement a generator function that reads a file line by line and yields each line as a string

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius_temp = [0, 10, 20, 30, 40]

def celcius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

fahrenheit_temp = list(map(lambda c: (c * 9/5) + 32, celsius_temp))

print(list(fahrenheit_temp))
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def is_not_vowel(char):
  vowels = "aeiouAEIOU"
  return char not in vowels

input_string = "Hello there!"

filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)
```

```python
# 11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

# Order Number      Book title and author               Quantity       Price per item
# 34587             Learning Python, Mark Lutz             4               40.95
# 98762             Programming Python, Mark Lutz          5               56.80
# 77226             Head First Python, Paul Berry          3               32.95
# 88112             Einfuhrung in Python3, Bernd Klein     3               24.99

#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#order is smaller than 100,00 €.

#Write a Python program using lambda and map.

orders = [
    [34587, 'Learning Python, Mark Lutz', 4, 40.95],
    [98762, 'Programming Python, Mark Lutz', 5, 56.80],
    [77226, 'Head First Python, Paul Berry', 3, 32.95],
    [88112, 'Einfuhrung in Python3, Bernd Klein', 3, 24.99]
]


def calculate_total(order):
    order_number, _, quantity, price_per_item = order
    total = quantity * price_per_item
    if total < 100:
        total += 10
    return (order_number, total)

result = list(map(lambda order: calculate_total(order), orders))

print(result)


```
